```python
import pandas as pd
from pathlib import Path
```

```python
raw_data_path= Path("../data/raw")
datasets={}
csv_files = list(raw_data_path.glob("*.csv"))
print(f"Found {len(csv_files)} CSV files")
for file in csv_files:
    try:
        df=pd.read_csv(file)
        dataset_name=file.stem
        datasets[dataset_name]=df
        print(f"loaded: {dataset_name}")
    except Exception as e:
        print(f"Error loading {file.name}:{e}")
```

```python
for name, df in datasets.items():

    print("=" * 80)
    print(f"DATASET: {name}")
    print("=" * 80)

    print("\nShape:")
    print(df.shape)

    print("\nColumn Names:")
    print(df.columns.tolist())

    print("\nData Types:")
    print(df.dtypes)

    print("\nFirst 5 Rows:")
    display(df.head())
```

```python
#missing values
for name, df in datasets.items():

    print(f"\n{name}")
    print(df.isnull().sum())
```

```python
#duplicate rows
for name, df in datasets.items():

    print(f"{name}: {df.duplicated().sum()} duplicate rows")
```

```python
#summary dataset
for name, df in datasets.items():

    print(f"{name}")
    print(f"Rows: {df.shape[0]}")
    print(f"Columns: {df.shape[1]}")
    print("-" * 40)
```

```python
#exploring fund_master
fund_master =pd.read_csv("../data/raw/01_fund_master.csv")
print("="*80)
print ("FUND MASTER EXPLORATION")
print("="*80)
# Shape
print("\nShape of Dataset:")
print(fund_master.shape)

# Data Types
print("\nData Types:")
print(fund_master.dtypes)

# First 5 Rows
print("\nFirst 5 Rows:")
print(fund_master.head())

# Missing Values
print("\nMissing Values:")
print(fund_master.isnull().sum())

# Duplicate Rows
print("\nDuplicate Rows:")
print(fund_master.duplicated().sum())

# Column Names
print("\nColumns:")
print(fund_master.columns.tolist())

# Descriptive Statistics
print("\nNumerical Summary:")
print(fund_master.describe())

# Check if Scheme Code is Unique
if 'scheme_code' in fund_master.columns:
    print("\nScheme Code Uniqueness:")
    print(fund_master['scheme_code'].is_unique)

    print("\nDuplicate Scheme Codes:")
    print(fund_master['scheme_code'].duplicated().sum())

# Unique Fund Houses
if 'fund_house' in fund_master.columns:
    print("\nNumber of Fund Houses:")
    print(fund_master['fund_house'].nunique())

    print("\nFund Houses:")
    print(fund_master['fund_house'].unique())

# Categories
if 'category' in fund_master.columns:
    print("\nCategory Distribution:")
    print(fund_master['category'].value_counts())

# Sub Categories
if 'sub_category' in fund_master.columns:
    print("\nSub Category Distribution:")
    print(fund_master['sub_category'].value_counts())

# Risk Grades
if 'risk_grade' in fund_master.columns:
    print("\nRisk Grade Distribution:")
    print(fund_master['risk_grade'].value_counts())
```

```python
import pandas as pd

fund_master = pd.read_csv("../data/raw/01_fund_master.csv")
nav_history = pd.read_csv("../data/raw/02_nav_history.csv")

# Get unique AMFI codes
master_codes = set(fund_master["amfi_code"])
nav_codes = set(nav_history["amfi_code"])

# Find missing codes
missing_codes = master_codes - nav_codes

print("Total AMFI Codes in fund_master:", len(master_codes))
print("Total AMFI Codes in nav_history:", len(nav_codes))
print("Missing AMFI Codes:", len(missing_codes))

if missing_codes:
    print("\nMissing AMFI Codes:")
    print(sorted(list(missing_codes)))
else:
    print("\n All AMFI codes in fund_master exist in nav_history")
```
